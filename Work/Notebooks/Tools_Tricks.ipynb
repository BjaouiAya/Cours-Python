## Functions

```python
def convert_to_datetimestring(step, start):
    """
        This parser convert into a SolisArt software format
    """
    for el in step:
        yield datetime.datetime.strftime(start +
                                         datetime.timedelta(seconds=int(el)),
                                         "%d/%m/%Y %H:%M")


def convert_to_datetime(step, start=datetime.datetime(2014, 1, 1)):
    """
      This parser convert to a real datetime format
    """
    for el in step:
        yield start + datetime.timedelta(seconds=int(el))


def convert_solis_to_datetime(solis_date):
    """
        This parseur convert SolisArt software format to a real datetime format
        input : 'day/month/year hour:minute'
        output : 'year-month-day hour:minute:second'
    """
    solis_date, hour = solis_date.split(" ")
    year, month, day = map(int, reversed(solis_date.split("/")))
    hour, minute = map(int, hour.split(":"))
    yield datetime.datetime(year, month, day, hour, minute)

    
def data_steps_iterator(dataframe, start=0, steps=[], interval=None):
    """
        Create an iterator to treat a dataframe by parts
        
        example :
            start=0
            steps=[el*48 for el in (31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31)]
            interval=12
        --> Month dataframe iterator
    """
    end = start + steps[0]
    yield dataframe[start:end]
    for el in range(interval-1):
        start = end
        end += steps[el+1]
        yield dataframe[start:end]
        
        
def set_boxes(box, colors=('#268bd2', '#002b36', '#268bd2', '#268bd2', '#dc322f'),
                   linewidth=(1.5, 1, 1, 0.5, 1), marker=''): 
    """
        Change boxplot color structure
        current display colors are : ('blue', 'black', 'blue', 'blue', 'red')
    """
    plt.setp(box['boxes'], color=colors[0], linewidth=linewidth[0])  # boxes 
    plt.setp(box['caps'], color=colors[1], linewidth=linewidth[1])  # limits 
    plt.setp(box['whiskers'], color=colors[2], linewidth=linewidth[2])  # points 
    plt.setp(box['fliers'], color=colors[3], linewidth=linewidth[3],
             marker=marker)  # markers
    plt.setp(box['medians'], color=colors[4], linewidth=linewidth[4])  # medians 
```

## Loading dataframe from csv

###### Load, change index, resample, sorting with timestep index

```python
# Here select folder
%cd "D:\GitHub\SolarSystem\Outputs\clean"
# Here csv
chambery_csv = 'chambery_30062014.csv'
# Your Dataframe structure
data_test = pd.read_csv(chambery_csv, skiprows=1, delimiter=";", index_col="Date")
# Convert index to datetime index
data_test.index = [ind for ind in convert_to_datetime(data_test.index)]
# Removes duplicated values
data_test = data_test.groupby(data_test.index).last()
# Resample to keep only one value each 30min (reduce data treatment and increase speed)
# Interpolate use to fill values
data_test = data_test[:].resample('30min').interpolate()
# Month length variation
months = (31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31)
# Store dataframe values of each months inside a list for T1 and Thouse
# el*48 used because datas sample = 30min ---> 48 values per day
T1_Tint = [el for el in data_steps_iterator(data_test[['T1', 'T12_house']],
                                               steps=[el*48 for el in months],
                                               interval=12)]
# Keep only daily sun values for each datframe inside previous list (8h to 18h30)
for i in range(len(T1_Tint)):
    T1_Tint[i] = T1_Tint[i].loc[T1_Tint[i].index.hour <= 18] # Will match 18 with 0 and 30 min
    T1_Tint[i] = T1_Tint[i].loc[T1_Tint[i].index.hour >= 9]
```
