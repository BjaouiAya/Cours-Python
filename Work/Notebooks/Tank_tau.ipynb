```python
# These parsers convert to a real datetime format

def convert_solis_to_datetime(solis_date):
    """
        This parseur convert SolisArt software format to a real datetime format
        input : 'day/month/year hour:minute'
        output : 'year-month-day hour:minute:second'
    """
    try:
        solis_date, hour = solis_date.split(" ")
        year, month, day = map(int, reversed(solis_date.split("/")))
        hour, minute = map(int, hour.split(":"))
        return datetime.datetime(year+1, month, day, hour, minute)
    # Bad csv with some text inside values --> return None to remove them with groupby later
    except ValueError:
        return None

    
def convert_to_datetime(step, start=datetime.datetime(2014, 1, 1)):
    """
      This parser convert to a real datetime format
    """
    for el in step:
        yield start + datetime.timedelta(seconds=int(el))
```

```python
simu = pd.read_csv("SeptOct_30_120_clean.csv", skiprows=1, delimiter=";", index_col="Date")
simu.index = [ind for ind in convert_to_datetime(simu.index)]
simu = simu.groupby(simu.index).last()
type(simu.index)
```

```python
# Csv columns names
for i, col in enumerate(simu.columns):
    print("column {}\t\t{}\t\t{}".format(i, col, type(simu[col][0])))
```

```python
# Create dataframe for september and october 
sept = simu[:]['2014-09-01 00:00:00':'2014-09-30 23:59:00']
octo = simu[:]['2014-10-01 00:00:00':'2014-10-31 23:59:00']
# Change sampling to match solisart sampling
sept = sept.resample('1min').interpolate()
octo = octo.resample('1min').interpolate()
```

```python
# Create new row and change it's index according to larger dataframe (sept_base or oct_base)
if len(sept) == len(sept_base)-1:
    add = sept.ix[len(sept)-1:, :]
    # Have to use "dataframe.ix" to get a datatime class and not a timestamp class
    add.index = sept_base.ix[:, :][len(sept_base)-1:].index
    sept = pd.concat([sept, add])
    print("September", len(sept), len(sept_base))
if len(octo) == len(oct_base)-1:    
    add = octo.ix[len(octo)-1:, :]
    # Have to use "dataframe.ix" to get a datatime class and not a timestamp class
    add.index = oct_base.ix[:, :][len(oct_base)-1:].index
    octo = pd.concat([sept, add])
    print("October", len(octo), len(oct_base))
else:
    print("Nothing to change")
```
