#### Prepare datetime formatter

```python
# Start time for timestep
year, month, day = 2014, 1, 1
start = datetime.datetime(year, month, day)
start
```

```python
# This parser convert to a real datetime format
def convert_to_datetime(step):
    for el in step:
        yield start + datetime.timedelta(seconds=int(el))
```

```python
# This parser convert into a solisart software format      
def convert_to_datetimestring(step):
    for el in step:
        yield datetime.datetime.strftime(start + datetime.timedelta(seconds=int(el+0.5)), "%d/%m/%Y %H:%M")
```

## --- Prepare dymola output datas ---

#### Read file datas

```python
test_plot = pd.read_csv("bad.csv", skiprows=1, delimiter=",", names=("Date", "T1", "T3", "T5"), index_col="Date")
test_solis = pd.read_csv("bad2.csv", skiprows=None, nrows=None, delimiter=",", index_col="Time")
```

```python
test_solis.columns = ['T1', 'T3', 'T5']
test_solis.index.name = "Date"
test_solis
```

#### Remove duplicates

```python
# Delete all duplicate time step
before = (len(test_plot.index), len(test_solis.index))
test_plot = test_plot.groupby(test_plot.index).last()
test_solis = test_solis.groupby(test_solis.index).last()
after = (len(test_plot.index), len(test_solis.index))
print("before:", before, "\t after:", after)
```

#### Convert columns values according to their fields. 

```python
# Cast kelvins to celsius
# Regular expression to catch column fields and do some modifications on them
reg = (re.compile("\AT\d+"), re.compile("\AHDi[fr]+"))
for column in test_plot.columns:
    if reg[0].match(column):
        test_plot[column] -= 273.15
    elif reg[1].match(column):
        test_plot[column] = 0
for column in test_solis.columns:
    if reg[0].match(column):
        test_solis[column] -= 273.15
    elif reg[1].match(column):
        test_solis[column] = 0
```
